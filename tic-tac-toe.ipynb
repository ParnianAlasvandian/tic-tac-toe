# Tic Tac Toe Game (Python)

A Python-based console implementation of the classic Tic Tac Toe game. The project demonstrates structured programming, game logic design, and interactive user input handling.

This function renders the Tic Tac Toe board in a 3x3 format. It clears the previous output to simulate a real-time update in the console and displays the current game state based on player moves.

```python
from IPython.display import clear_output

def display_board(board):
    clear_output()
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
```

```python
test_board = ['#','X','O','X','O','X','O','X','O','X']
display_board(test_board)
```

This function handles player input for selecting markers in the game. Player 1 chooses either 'X' or 'O', and Player 2 is automatically assigned the opposite marker. The function ensures valid input before proceeding.

```python
def player_input():
    '''
    OUTPUT: (Player 1 marker, Player 2 marker)
    '''

    choice = ''

    while choice not in ['X', 'O']:
        choice = input('Player 1, choose X or O: ').upper()

        if choice not in ['X', 'O']:
            print('Sorry, invalid choice!')

    if choice == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')
```

This function places a player's marker ('X' or 'O') on the board at the specified position. It updates the board state based on the player's move.

```python
def place_marker(board, marker, position):
    board[position] = marker
```

This function determines if a player has won the game by checking all possible winning conditions on the Tic Tac Toe board, including horizontal, vertical, and diagonal patterns.

```python
def win_check(board, mark):

    return ((board[1] == board[2] == board[3] == mark) or
            (board[4] == board[5] == board[6] == mark) or
            (board[7] == board[8] == board[9] == mark) or
            (board[1] == board[4] == board[7] == mark) or
            (board[2] == board[5] == board[8] == mark) or
            (board[3] == board[6] == board[9] == mark) or
            (board[1] == board[5] == board[9] == mark) or
            (board[3] == board[5] == board[7] == mark))
```

This function randomly determines which player will go first by simulating a coin flip using Python’s random module.

```python
import random

def choose_first():

    flip = random.randint(0,1)

    if flip == 0:
        return 'Player 1'
    else:
        return 'Player 2'
```

This function checks whether a specific position on the board is available for a move by verifying if it is still unoccupied.

```python
def space_check(board, position):
    
    return board[position] in ['1' , '2' , '3' , '4' , '5' , '6' , '7' , '8' , '9']
```

This function checks if the board is completely filled. If no empty spaces remain, the board is considered full and the game ends in a draw if no winner is found.

```python
def full_board_check(board):

    for i in range(1,10):
        if space_check(board,i):
            return False

    # BOARD IS FULL IF WE RETURN True
    return True
```

This function prompts the player to select a valid position on the board. It ensures the input is within the valid range and that the chosen space is not already occupied.

```python
def player_choice(board):

    position = 0

    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input(' Choose a position: (1-9) '))
        
    return position    
```

This function asks the player if they would like to play another game and returns a boolean value based on their response.

```python
def replay():

    choice = input('Play again? Enter Yes or No')

    return choice == 'Yes'
```

## Main Game Loop

This is the core game engine of the project. It manages the full game flow, including setup, alternating player turns, state updates, win/draw detection, and replay control.

```python
print('Welcome to Tic Tac Toe!')

while True:

    ## SET EVERYTHING UP (BOARD , WHOSE FIRST , CHOOSE MARKERS X,O)
    the_board = ['#', '1' , '2' , '3' , '4' , '5' , '6' , '7' , '8' , '9']
    player1_marker,player2_marker = player_input()

    turn = choose_first()
    print(turn + ' will go first.')

    play_game = input('Ready to play? y or n?')

    if play_game.lower() == 'y':
        game_on = True
    else:
        game_on = False

    ## GAME PLAY
    while game_on:
        if turn == 'Player 1':
            
        ### PLAYER ONE TURN
            # show the board
            display_board(the_board)
            
            # choose a position
            print(f'Player 1  with {player1_marker} => ')
            position = player_choice(the_board)

            # place the marker on the position
            place_marker(the_board, player1_marker, position)

            # check if the won
            if win_check(the_board, player1_marker):
                display_board(the_board)
                print(' PLAYER 1 HAS WON!!!')
                game_on = False
            else:    
            # check if there is a tie
                if full_board_check(the_board):
                    display_board(the_board)
                    print(' TIE GAME!')
                    game_on = False
                else:
                    turn = 'Player 2'
            # No tie and no win? Then next player's turn
        
        else:
            
        ### PLAYER TWO TURN
            # show the board
            display_board(the_board)
            
            # choose a position
            print(f'Player 2  with {player2_marker} => ')
            position = player_choice(the_board)

            # place the marker on the position
            place_marker(the_board, player2_marker, position)

            # check if the won
            if win_check(the_board, player2_marker):
                display_board(the_board)
                print(' PLAYER 2 HAS WON!!!')
                game_on = False
            else:    
            # check if there is a tie
                if full_board_check(the_board):
                    display_board(the_board)
                    print(' TIE GAME!')
                    game_on = False
                else:
                    turn = 'Player 1'


    if not replay():
        break
```
